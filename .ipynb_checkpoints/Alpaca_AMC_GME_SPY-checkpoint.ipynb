```python
# Import libraries and env
import os
import requests
import pandas as pd
from dotenv import load_dotenv
import alpaca_trade_api as tradeapi

%matplotlib inline
load_dotenv()
```

```python
# Alpaca keys & APIs
alpaca_API = os.getenv("alpaca_API")
alpaca_secret_key = os.getenv("alpaca_secret_key")
```

```python
api = tradeapi.REST(
    alpaca_API,
    alpaca_secret_key,
    api_version="v2")

# Get daily price data for GME over the last 100 days.
barset = api.get_barset('GME', 'day')
start = pd.Timestamp("2021-01-01", tz="America/New_York").isoformat()
end = pd.Timestamp("2021-05-26", tz="America/New_York").isoformat()
tickers = ["GME", "AMC", "SPY"]
timeframe = "1D"

# Get current closing prices for GME, AMC and SPY
df_meme_stonks = api.get_barset(
    tickers,
    timeframe,
    start = start,
    end = end
).df

# Display sample data
df_meme_stonks

##gme_bars = barset['GME']

# See how much GME moved in that timeframe.
#week_open = gme_bars[0].o
#week_close = gme_bars[-1].c
#percent_change = (week_close - week_open) / week_open * 100
#print('GME moved {}% over the last 100 days'.format(percent_change))
```

```python
# Create and empty DataFrame for closing prices
df_meme_close = pd.DataFrame()

# Fetch the closing prices of FB and TWTR
df_meme_close["AMC"] = df_meme_stonks["AMC"]["close"]
df_meme_close["GME"] = df_meme_stonks["GME"]["close"]
df_meme_close["SPY"] = df_meme_stonks["SPY"]["close"]

# Drop the time component of the date
df_meme_close.index = df_meme_close.index.date

# Display sample data
df_meme_close
```
